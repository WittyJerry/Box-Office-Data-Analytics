## Task 2: Visualizing the Target Distribution

```python
#Visualizing the revenue distribution
sns.distplot (df['revenue'], color='blue')
plt.title ('Revenue')
plt.show()
```

```python
#visualizing the log transformed revenue distribution
sns.distplot (np.log1p(df['revenue']), color='red')
plt.title ('Log Transformed Revenue')
plt.show()
```

## Task 3: Relationship between Film Revenue and Budget

```python
#View first 10 rows in the revenue and budget columns
print ("The first 10 rows in the revenue and budget columns")
df[['revenue', 'budget']]. head(10)
```

```python
#Visualize film revenue and budget
sns.scatterplot (df['revenue'], df['budget'])
plt.title ("Relationship between Film Revenue and Budget")
plt.show()
```

```python
#Visualize relationship between film revenue and budget
sns.regplot(df['revenue'], df['budget'])
plt.title ("Relationship between Film Revenue and Budget")
plt.show()
```

```python
#visualize Log transformed Revenue and budget
sns.scatterplot(np.log1p(df['revenue']), df['budget'])
plt.title ("Log Transformed Film Revenue and Budget")
plt.show()
```

```python
#Visualize the relationship between Log Transformed Revenue and budget
sns.regplot(np.log1p(df['revenue']), df['budget'])
plt.title ("Log Transformed Film Revenue and Budget")
plt.show()
```

```python
Relationship= df[['revenue','budget']].corr()
sns.heatmap (Relationship, vmin=0, vmax=1, center=0)
plt.title ("Relationship between Film Revenue and Budget")
plt.show()
```
